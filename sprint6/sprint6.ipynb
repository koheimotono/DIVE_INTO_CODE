```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
class ScratchDecesionTreeClassifierDepth1():
    """
    深さ1の決定木分類器のスクラッチ実装

    Parameters
    ----------
    verbose : bool
      学習過程を出力する場合はTrue
    """

    def __init__(self, verbose=False):
        # ハイパーパラメータを属性として記録
        self.verbose = verbose
        
        self.labels = None
        self.max_gain_info = None
        self.max_gain = 0
        self.left = None
        self.right = None
        self.left_label = None
        self.right_label = None

    def fit(self, X, y):
        """
        決定木分類器を学習する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            学習用データの正解値
        """
        self.labels = np.unique(y)
        self.node_split(X, y)
        
        if self.verbose:
            self.area(X, y)
            print()

    def predict(self, X):
        """
        決定木分類器を使いラベルを推定する
        """
        threshold, i_feature = self.max_gain_info
        return np.where(X[:, i_feature] >= threshold, self.right_label, self.left_label)
    
    def node_split(self, X, y):
        """
        ノード間の 情報利得 を計算する関数

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
                学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
                学習用データの正解値
        """
        n_samples, n_features = X.shape
        self.max_gain = 0
        for i in range(n_features):
            Xi = X[:, i]
            values = np.unique(Xi)
            for threshold in values[1:]:
                idx = np.arange(n_samples)
                left, right = idx[Xi < threshold], idx[Xi >= threshold]
                gain_temp = gain(y, left, right)
                if self.max_gain < gain_temp:
                    self.max_gain = gain_temp
                    self.max_gain_info = [threshold, i]
                    self.left, self.right = left, right
        self.labeling(y, self.left, self.right)
        
    def labeling(self, y, left, right):
        y1, y2 = self.labels
        right_y1, right_y2 = np.count_nonzero(y[right]==y1), np.count_nonzero(y[right]==y2)
        left_y1, left_y2 = np.count_nonzero(y[left]==y1), np.count_nonzero(y[left]==y2)
        if right_y1*left_y2 > left_y1*right_y2:
            self.right_label, self.left_label = y1, y2
        else:
            self.left_label, self.right_label = y1, y2
    
    def area(self, X, y):
        print("max_gain_info =", self.max_gain_info)
        plt.plot(X[y==self.labels[0],0], X[y==self.labels[0],1], "o", label="y1")
        plt.plot(X[y==self.labels[1],0], X[y==self.labels[1],1], "x", label="y2")
        if self.max_gain_info[1] == 0:
            plt.axvline(self.max_gain_info[0], color = "magenta")
        else:
            plt.axhline(self.max_gain_info[0], color = "magenta")
        plt.legend()
        plt.show()
```

## 【問題1】不純度を求める関数

```python
def gini_coef(y):
    """
    ノード の ジニ不純度 を計算する関数
    
    Parameters
    ----------
    y_i : i番目のクラス
    """
    sigma = 0
    for label in np.unique(y):
        sigma += (np.count_nonzero(y==label))**2
    sigma /= (len(y))**2
    return 1 - sigma
```

## 【問題2】情報利得を求める関数

```python
def gain(y, left, right):
    """
    ノード間の 情報利得 を計算する関数
    
    Parameters
    ----------
    y_i : i番目のクラス
    left : 左側のノードを示すインデックス
    right : 右側のノードを示すインデックス
    """
    return gini_coef(y) - (len(left)/len(y))*gini_coef(y[left]) - (len(right)/len(y))*gini_coef(y[right])
```

## 【問題5】学習と推定

```python
def scratch_train_test_split(X, y, train_size=0.8,):
    """
    検証用データを分割する。

    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      学習データ
    y : 次の形のndarray, shape (n_samples, )
      正解値
    train_size : float (0<train_size<1)
      何割をtrainとするか指定

    Returns
    ----------
    X_train : 次の形のndarray, shape (n_samples, n_features)
      学習データ
    X_test : 次の形のndarray, shape (n_samples, n_features)
      検証データ
    y_train : 次の形のndarray, shape (n_samples, )
      学習データの正解値
    y_test : 次の形のndarray, shape (n_samples, )
      検証データの正解値
    """
    n_samples = len(X)
    idx = np.zeros(n_samples)
    train_size =  int(n_samples*train_size)
    train_idx = np.random.choice(np.arange(n_samples), train_size, replace=False)
    idx[train_idx] = 1
    
    X_train ,X_test = X[idx==1], X[idx==0]
    y_train ,y_test = y[idx==1], y[idx==0]
    
    return X_train, X_test, y_train, y_test
```

```python
def evaluate(y_true, y_pred):
    print("accuracy =", accuracy_score(y_true, y_pred))
    print("precision =", precision_score(y_true, y_pred, average='macro'))
    print("recall =", recall_score(y_true, y_pred, average='macro'))
    print("f1 =", f1_score(y_true, y_pred, average='macro'))
    print(confusion_matrix(y_true, y_pred))
```

```python
def decision_region(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['y1', 'y2']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue', 'yellow']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
def learn_predict(X, y, model, print_y_pred=False):
    X_train, X_test, y_train, y_test = scratch_train_test_split(X, y)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    if print_y_pred:
        print("y_pred =", y_pred)

    decision_region(X_train, y_train, model)
    decision_region(X_test, y_test, model)
    evaluate(y_pred, y_test)

def learn(X, y, model, print_y_pred=False):
    model.fit(X, y)
    decision_region(X, y, model)
    y_pred = model.predict(X)
    if print_y_pred:
        print("y_pred =", y_pred)
    evaluate(y_pred, y)
    #model.tree
```

## 【問題7】（アドバンス課題）深さ2の決定木分類器クラスの作成

```python
class ScratchDecesionTreeClassifierDepth2():
    """
    深さ1の決定木分類器のスクラッチ実装

    Parameters
    ----------
    verbose : bool
      学習過程を出力する場合はTrue
    """

    def __init__(self, verbose=False):
        # ハイパーパラメータを属性として記録
        self.verbose = verbose
        
        self.labels = None
        self.tree = None
        self.y = None

    def fit(self, X, y):
        """
        決定木分類器を学習する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            学習用データの正解値
        """
        self.y = y
        
        self.labels = np.unique(y)
        idx = np.arange(X.shape[0])
        self.tree = self.node_split(X, y, idx) #1回目の分岐
        idx_left, idx_right = self.tree[:2]
        self.tree = [self.node_split(X[idx_left], y[idx_left], idx_left), self.node_split(X[idx_right], y[idx_right], idx_right), self.tree[2]] #2回目の分岐
        
        if self.verbose:
            self.area(X, y)
            print()

    def predict(self, X):
        """
        決定木分類器を使いラベルを推定する
        """
        y_pred = np.empty(X.shape[0])
        for i in range(2):
            for j in range(2):
                s = 1 if i==0 else -1
                t = 1 if j==0 else -1
                y_pred[np.where((s * (X[:, self.tree[2][1]] - self.tree[2][0]) <= 0) & (t * (X[:, self.tree[i][2][1]] - self.tree[i][2][0]) <= 0))] = self.labeling(self.y[self.tree[i][j]], self.y)
        return y_pred
    
    def node_split(self, X, y, idx):
        """
        ノード間の 情報利得 を計算する関数

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
                学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
                学習用データの正解値
        """
        n_samples, n_features = X.shape
        max_gain = 0
        for i in range(n_features):
            Xi = X[:, i]
            values = np.unique(Xi)
            for threshold in values[1:]:
                left, right = idx[Xi < threshold], idx[Xi >= threshold]
                gain_temp = gain(self.y, left, right)
                if max_gain < gain_temp:
                    max_gain = gain_temp
                    max_gain_info = [left, right, [threshold, i]]
        return max_gain_info
        
    def labeling(self, y, y_all):
        y1, y2 = self.labels
        count_y1, count_y2 = np.count_nonzero(y==y1), np.count_nonzero(y==y2)
        all_y1, all_y2 = np.count_nonzero(y_all==y1), np.count_nonzero(y_all==y2)
        if count_y1*all_y2 > count_y2*all_y1:
        #if count_y1 >= count_y2:
            return y1
        else:
            return y2
```

## 【問題8】（アドバンス課題）深さに制限のない決定木分類器クラスの作成

```python
class ScratchDecesionTreeClassifierDepthInf():
    """
    深さ1の決定木分類器のスクラッチ実装

    Parameters
    ----------
    verbose : bool
      学習過程を出力する場合はTrue
    """

    def __init__(self, max_depth=-1, verbose=False):
        # ハイパーパラメータを属性として記録
        self.max_depth = max_depth
        self.verbose = verbose
        
        self.labels = None
        self.tree = None
        self.y = None
        self.y_pred = None

    def fit(self, X, y):
        """
        決定木分類器を学習する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            学習用データの正解値
        """
        self.y = y
        
        self.labels = np.unique(y)
        idx = np.arange(X.shape[0])
        self.tree = self.node_split(X, y, idx)
        self.tree = self._split(X, self.tree)
        
        if self.verbose:
            self.area(X, y)
            print()
    
    def _split(self, X, node_p, i=1):
        idx_left, idx_right = node_p[:2]
        node_lc, node_rc = node_p[:2]
        if i != self.max_depth:
            if gini_coef(self.y[node_p[0]]) > 0:
                node_lc = self._split(X, self.node_split(X[idx_left], self.y[idx_left], idx_left), i+1)
            if gini_coef(self.y[node_p[1]]) > 0:
                node_rc = self._split(X, self.node_split(X[idx_right], self.y[idx_right], idx_right), i+1)
        return [node_lc, node_rc, node_p[2]]
    
    def predict(self, X):
        """
        決定木分類器を使いラベルを推定する
        """
        self.y_pred = np.empty(X.shape[0])
        bl_idx = np.full(X.shape[0], True)
        self._predict(X, self.tree, bl_idx)
        return self.y_pred
    
    def _predict(self, X, node_p, bl_idx):
        for j in range(2):
            t = 1 if j==0 else -1
            bl_idx1 = bl_idx & (t * (X[:, node_p[2][1]] - node_p[2][0]) <= 0)
            if type(node_p[j][-1]) == list:
                self._predict(X, node_p[j], bl_idx1)
            else:
                self.y_pred[np.where(bl_idx1)] = self.labeling(self.y[node_p[j]], self.y)
    
    def node_split(self, X, y, idx):
        """
        ノード間の 情報利得 を計算する関数

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
                学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
                学習用データの正解値
        """
        n_samples, n_features = X.shape
        max_gain = 0
        for i in range(n_features):
            Xi = X[:, i]
            values = np.unique(Xi)
            for threshold in values[1:]:
                left, right = idx[Xi < threshold], idx[Xi >= threshold]
                gain_temp = gain(self.y, left, right)
                if max_gain < gain_temp:
                    max_gain = gain_temp
                    max_gain_info = [left, right, [threshold, i]]
        return max_gain_info
        
    def labeling(self, y, y_all):
        y1, y2 = self.labels
        count_y1, count_y2 = np.count_nonzero(y==y1), np.count_nonzero(y==y2)
        all_y1, all_y2 = np.count_nonzero(y_all==y1), np.count_nonzero(y_all==y2)
        if count_y1*all_y2 > count_y2*all_y1:
        #if count_y1 >= count_y2:
            return y1
        else:
            return y2
```

## 【その他】深さに制限のない決定木分類器クラス（境界線は点の間）

```python
class ScratchDecesionTreeClassifierDepthInf2():
    """
    深さ1の決定木分類器のスクラッチ実装

    Parameters
    ----------
    verbose : bool
      学習過程を出力する場合はTrue
    """

    def __init__(self, max_depth=-1, verbose=False):
        # ハイパーパラメータを属性として記録
        self.max_depth = max_depth
        self.verbose = verbose
        
        self.labels = None
        self.tree = None
        self.y = None
        self.y_pred = None

    def fit(self, X, y):
        """
        決定木分類器を学習する
        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            学習用データの正解値
        """
        self.y = y
        
        self.labels = np.unique(y)
        idx = np.arange(X.shape[0])
        self.tree = self.node_split(X, y, idx) # 決定木を保存するリスト [left_child, right_child, split_info]
        self.tree = self._split(X, self.tree)
        
        if self.verbose:
            self.area(X, y)
            print()
    
    def _split(self, X, node_p, i=1):
        idx_left, idx_right = node_p[:2]
        node_lc, node_rc = node_p[:2]
        if i != self.max_depth:
            if gini_coef(self.y[node_p[0]]) > 0:
                node_lc = self._split(X, self.node_split(X[idx_left], self.y[idx_left], idx_left), i+1)
            if gini_coef(self.y[node_p[1]]) > 0:
                node_rc = self._split(X, self.node_split(X[idx_right], self.y[idx_right], idx_right), i+1)
        return [node_lc, node_rc, node_p[2]]
    
    def predict(self, X):
        """
        決定木分類器を使いラベルを推定する
        """
        self.y_pred = np.empty(X.shape[0])
        bl_idx = np.full(X.shape[0], True)
        self._predict(X, self.tree, bl_idx)
        return self.y_pred
    
    def _predict(self, X, node_p, bl_idx):
        for j in range(2):
            t = 1 if j==0 else -1
            bl_idx1 = bl_idx & (t * (X[:, node_p[2][1]] - node_p[2][0]) <= 0)
            if type(node_p[j][-1]) == list:
                self._predict(X, node_p[j], bl_idx1)
            else:
                self.y_pred[np.where(bl_idx1)] = self.labeling(self.y[node_p[j]], self.y)
    
    def node_split(self, X, y, idx):
        """
        ノード間の 情報利得 を計算する関数

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
                学習用データの特徴量
        y : 次の形のndarray, shape (n_samples, )
                学習用データの正解値
        """
        n_samples, n_features = X.shape
        max_gain = 0
        for i in range(n_features):
            Xi = X[:, i]
            values = np.unique(Xi)
            for j in range(len(values) - 1):
                threshold = (values[j+1] + values[j]) / 2
                left, right = idx[Xi < threshold], idx[Xi >= threshold]
                gain_temp = gain(self.y, left, right)
                if max_gain < gain_temp:
                    max_gain = gain_temp
                    max_gain_info = [left, right, [threshold, i]]
        return max_gain_info
        
    def labeling(self, y, y_all):
        y1, y2 = self.labels
        count_y1, count_y2 = np.count_nonzero(y==y1), np.count_nonzero(y==y2)
        all_y1, all_y2 = np.count_nonzero(y_all==y1), np.count_nonzero(y_all==y2)
        if count_y1*all_y2 > count_y2*all_y1:
        #if count_y1 >= count_y2:
            return y1
        else:
            return y2
```
